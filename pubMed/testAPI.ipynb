```python
from Bio import Entrez
import pandas as pd
import numpy as np
```

```python
def search(query):
    Entrez.email = 'email@example.com'
    handle = Entrez.esearch(db='pubmed',
                           sort='relevance',
                           retmax='25',
                           retmode='xml',
                           term=query)
    results = Entrez.read(handle)
    return results

studies = search('COVID-19')
studiesIdList = studies['IdList']

```

```python
import metapub

for id in studiesIdList:
    article = metapub.PubMedFetcher().article_by_pmid(id)
    if article.doi:
        article_link = 'https://doi.org/' + article.doi
        print(article_link)
```

```python
def fetch_details(id_list):
    ids = ','.join(id_list)
    Entrez.email = 'email@example.com'
    handle = Entrez.efetch(db='pubmed',
                          retmode='xml',
                          id=ids)
    results = Entrez.read(handle)
    return results
```

```python
papers = fetch_details(studiesIdList)
for paper in papers['PubmedArticle']:
    print(paper['MedlineCitation']['Article'])
    input()
```
